```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_data = mouse_drug_data.merge(clinical_trial_data, on='Mouse ID')

# Display the data table for preview
merge_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
grouped_Drug_Timepoint = merge_data.groupby(['Drug','Timepoint'], as_index=False)[['Tumor Volume (mm3)']].mean()

# Preview DataFrame
grouped_Drug_Timepoint.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
sem = merge_data.groupby(['Drug', 'Timepoint'])[['Tumor Volume (mm3)']].sem()
sem = sem.reset_index(level=['Drug', 'Timepoint'])

# Preview DataFrame
sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot = grouped_Drug_Timepoint.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
pivot.head()
```

```python
# Generate the Plot (with Error Bars)
errors = sem.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
my_plot = pivot.plot.line(yerr=errors, figsize=(10,10))
my_plot.set_ylabel('Tumor Volume (mm3)')

# Save the Figure
fig = my_plot.get_figure()
fig.savefig("Tumor Volume.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
MetSite = merge_data.groupby(['Drug','Timepoint'], as_index=False)[['Metastatic Sites']].mean()

# Preview DataFrame
MetSite.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
MetSiteError = merge_data.groupby(['Drug', 'Timepoint'])[['Metastatic Sites']].sem()
MetSiteError = MetSiteError.reset_index(level=['Drug', 'Timepoint'])

# Preview DataFrame
MetSiteError.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
MetSitePivot = MetSite.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

MetSitePivot.head()
```

```python
# Generate the Plot (with Error Bars)
errors = MetSiteError.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
my_plot = MetSitePivot.plot.line(yerr=errors, figsize=(10,10))
my_plot.set_ylabel('Metastatic Sites')

# Save the Figure
fig = my_plot.get_figure()
fig.savefig("Metastatic Sites.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
MouseID = merge_data.groupby(['Drug','Timepoint'], as_index=False)[['Mouse ID']].count()

# Preview DataFrame
MouseID.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
MousePivot = MouseID.pivot(index='Timepoint', columns='Drug', values='Mouse ID')

# Preview the Data Frame
MousePivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
MousePercent = MousePivot / list(MousePivot.max()) * 100
errors = [0.5] * len(MousePercent) #*******************************************************************************************

my_plot = MousePercent.plot.line(yerr=errors, figsize=(10,10), marker='s')
my_plot.set_ylabel('Survival Rate (%)')

# Save the Figure
fig = my_plot.get_figure()
fig.savefig("Survival Rate.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
first_row = pivot.iloc[0]
last_row = pivot.iloc[len(pivot)-1]
pct = (last_row - first_row) / first_row * 100

# Display the data to confirm
pct
```

```python
# Store all Relevant Percent Changes into a Tuple
tumor_tuple = list(zip(pct.index, pct))

# Splice the data between passing and failing drugs
check = [0] * len(tumor_tuple)

i = 0
for tumor in tumor_tuple:
    if tumor[1] < 0:
        check[i] = (tumor[0], tumor[1], 'green')
    else:
        check[i] = (tumor[0], tumor[1], 'red')
    i += 1

# Orient widths. Add labels, tick marks, etc. 
# Use functions to label the percentages of changes
# Call functions to implement the function calls
x, y, z = [0] * len(tumor_tuple), [0] * len(tumor_tuple), [0] * len(tumor_tuple)

i = 0
for tumor in check:
    x[i], y[i], z[i] = tumor[0], tumor[1], tumor[2]
    i += 1

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
plt.ylabel('% Tumor Volume Change')
plt.bar(x,y, color=z)
plt.title('Tumor Change over 45 Day Treatment')

# Save the Figure
plt.savefig('Tumor Change.png')

# Show the Figure
plt.show()
```
